```python
def show_board(board):
    '''
    This function prints a visual representation of the current board.
    '''
    
    space = ' | '
    index=0
    
    for row in board:
        print(" " + str(row[0]) + space + str(row[1]) + space + str(row[2]))
        if index == 0 or index == 1:
            print('-----------')
        index += 1
```

```python
def valid_move(position, board):
    '''
    This function makes sure that a potential move indicated by the player is a valid move. 
    '''
    
    switch = {
        1:(board[2][0] == 1),
        2:(board[2][1] == 2),
        3:(board[2][2] == 3),
        
        4:(board[1][0] == 4),
        5:(board[1][1] == 5),
        6:(board[1][2] == 6),
        
        7:(board[0][0] == 7),
        8:(board[0][1] == 8),
        9:(board[0][2] == 9)
    }
    
    if(position not in [1,2,3,4,5,6,7,8,9]):
        return False
    
    return switch[position]
        
```

```python
def play(board, position, team):
    """
    This function places team's piece at the position (1-9) on the board.
    """
    
    if position == 1:
        board[2][0] = team

    if position == 2:
        board[2][1] = team

    if position == 3:
        board[2][2] = team

    if position == 4:
        board[1][0] = team

    if position == 5:
        board[1][1] = team

    if position == 6:
        board[1][2] = team

    if position == 7:
        board[0][0] = team

    if position == 8:
        board[0][1] = team

    if position == 9:
        board[0][2] = team
    
    return board
```

```python
def end_state_achieved(board):
    """
    This function checks the state of the board and returns the winner. Returns player 1's icon, player 2's icon, None if noone has won yet, or 'tie' if noone has won and there are no more moves
    """
    
    # Check if tie
    count = 0
    for row in board:
        for col in row:
            if type(col)==int:
                count += 1
    if count == 0:
        return 'tie'
    
    
    # Check x and o wins
    for row in range(len(board)):

        # Three in a row row-wise
        if board[row][0] == board[row][1]:
            if board[row][1] == board[row][2]:
                return board[row][2]

        # Three in a row col-wise
        if board[0][row] == board[1][row]:
            if board[1][row] == board[2][row]:
                return board[2][row]
    
    # Diagonal \
    if board[0][0] == board[1][1]:
            if board[1][1] == board[2][2]:
                return board[2][2]
    
    # Diagonal /
    if board[2][0] == board[1][1]:
            if board[1][1] == board[0][2]:
                return board[0][2]
    
    
    return None
```

```python
# Main game

#ask player 1 initial
x_input = str(input("Input player 1's singular initial (Default: X): "))
if x_input == '':
    x = 'X'
else:
    x = x_input[0].upper()
print(f'Player 1 is {x}')
print()

#ask player 2 initial
o_input = str(input("Input player 2's singular initial (Default: O): "))
# X always gets their initial. Second player gets what they ask for as long as it doesn't conflict with x's initial
if o_input == '':
    if x == 'O': 
        o = 'X'
    else:
        o = 'O'
else:
    if o_input[0].upper() == x:
        if x == 'O': 
            o = 'X'
        else:
            o = 'O'
    else:
        o = o_input[0].upper()
print(f'Player 2 is {o}')
print()

# Instantiate the board and other global vars
current_board = [[7, 8, 9], [4, 5, 6], [1, 2, 3]]
turn = 1
curr_player = x
valid_game = True
running = True

# Make it play in the collapsed jupyter notebook nicely
print('\n\n\n\n\n\n\n\n\n\n')

while (running):
    # Check to make sure the game board is still valid
    try:
        test_state(current_board)
    except InvalidBoard as e:
        print(e)
        print('Please restart game.')
    
    # Alternate players
    curr_player = x if turn % 2 == 1 else o
        
    # Obtain player's input
    print(f"{curr_player}'s turn:")
    show_board(current_board)
    attempt = input('Where would you like to go? (Any of the available numbered spaces left): ')
    
    # Validate player input (number)
    try:
        attempt = int(attempt)
    except Exception as e:
        print(e)
        print("This was a human, wasn't it?")
        valid_game = False #human error is not going to be added to training cases for the NN
        break
        
    # Validate player input (position available)
    if not valid_move(attempt, current_board):
        print('Move was not valid, game over.')
        break
        # I am not allowing multiple entries for an invalid input because I want to capture invalid moves as losses for training the NN
    
    # Make the move
    current_board = play(current_board, attempt, curr_player)
    
    print('\n\n\n\n\n\n\n\n\n\n\n')
    
    # Check for end goal
    winner = end_state_achieved(current_board)
    if winner != None:
        running=False
        break    
    
    turn += 1

# print out the winning team, turn number, that sort of thing
# record information for NN training
show_board(current_board)
if not valid_game:
    print('Game invalid, do not use for NN training.')

if winner == 'tie':
    print(f'The game ended in a tie on turn {turn}.')
else:
    print(f'{winner} won on turn {turn}!')
```
